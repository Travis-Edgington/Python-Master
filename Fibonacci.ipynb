```python


def fibonacci(n):
        f0=0
        f1=1
        a=0
        b=1

        if n==1:
            return 0
        else:
            list=[f0, f1]
            while len(list) <= n:
                list.append(list[a]+ list[b])
                a+=1
                b+=1
            return list
          




```

```python
fibonacci(20)
```

```python
import matplotlib.pyplot as plt
# Plot the list
plt.plot(fibonacci(20), marker='o', linestyle='-', color='b')  # Add styling options
plt.title('Plot of List Elements')  # Title of the plot
plt.xlabel('Index')                 # Label for the x-axis
plt.ylabel('Value')                 # Label for the y-axis
plt.grid(True)                      # Add a grid
plt.show()   
```
